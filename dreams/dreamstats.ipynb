```python
import os
```

```python
from collections import Counter
```

```python
def tryint(s):
    try:
        return int(s)
    except ValueError:
        return -1
```

```python
year = Counter()
month = Counter()
day = Counter()
for f in os.listdir(os.getcwd()):
    if f.endswith('.md'):
        year[tryint(f[:4])] += 1
        month[tryint(f.split('-')[1])] += 1
        day[tryint(f.split('-')[2])] += 1
```

```python
month.pop(-1)
day.pop(-1)
```

```python
year.most_common()
```

```python
month.most_common()
```

```python
day.most_common()
```

```python
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.figure(figsize=(16,12))
xs = sorted(month.keys())
ys = [month[x] for x in xs]
plt.plot(xs, ys)
```

```python
plt.figure(figsize=(16,12))
xs = sorted(day.keys())
ys = [day[x] for x in xs]
plt.plot(xs, ys)
```
